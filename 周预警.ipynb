# 读取所有项目信息

```python
qysx = pd.read_excel(r'D:\raw_data\企业属性20200601.xlsx', dtype={'开户时间': 'datetime64', '切换系统时间': 'datetime64',
                                                              '首次建单时间': 'datetime64', '末次建单时间': 'datetime64',
                                                              '首次支付时间': 'datetime64', '末次支付时间': 'datetime64',
                                                              '首次开票时间': 'datetime64', '末次开票时间': 'datetime64'})
```

```python
# 筛选 集团下 的项目信息
data_jituan = qysx.query("部门 == '集团一部' or 部门 == '集团二部' or 部门 == '集团三部' or 部门 == '集团四部' \
                          or 部门 == '实施运营部' or 部门 == '商务运营部'").copy()
# 筛选 KA和区域 的项目信息
data_ka_quyu = qysx.query("部门 != '集团一部' and 部门 != '集团二部' and 部门 != '集团三部' and 部门 != '集团四部' \
                          and 部门 != '实施运营部' and 部门 != '商务运营部'").copy()
```

## 集团项目-基于企业ID-处理

```python
# 对集团下的 项目 按 企业ID 去重
data_jituan.sort_values(by=['企业ID', '首次建单时间'], ascending=True, inplace=True)
# 获取每个项目对应的部门，销售等基础信息
dept_sale_class_df = data_jituan.groupby('企业ID')[['企业名称', '受票方名称', '部门', '销售', '运营', '联系人',
                                                  '联系电话', '使用系统', '服务费率', '分级', '业务类型']].last()
dept_sale_class_df.reset_index(inplace=True)

# 获取每个项目最早开户时间最早建单时间最后建单时间
open_bo_time_df = data_jituan.groupby('企业ID').agg({'开户时间': 'min', '首次建单时间': 'min', '末次建单时间': 'max'})
open_bo_time_df.reset_index(inplace=True)
# 剔除首次支付为空的项目
data_jituan2 = data_jituan.dropna(subset=['首次支付时间'])
# 获取有支付行为项目最早支付时间
first_pay_time = data_jituan2.groupby('企业ID')['首次支付时间'].min()
first_pay_time_df = pd.DataFrame({'企业ID': first_pay_time.index, '首次支付时间': first_pay_time.values})
# 获取有支付行为项目最后支付时间
last_pay_time = data_jituan2.groupby('企业ID')['末次支付时间'].max()
last_pay_time_df = pd.DataFrame({'企业ID': last_pay_time.index, '末次支付时间': last_pay_time.values})
# 剔除首次开票为空的项目
data_jituan3 = data_jituan2.dropna(subset=['首次开票时间'])
# 获取有开票行为项目最早开票时间
first_bill_time = data_jituan3.groupby('企业ID')['首次开票时间'].min()
first_bill_time_df = pd.DataFrame({'企业ID': first_bill_time.index, '首次开票时间': first_bill_time.values})
# 获取有开票行为项目最后开票时间
last_bill_time = data_jituan3.groupby('企业ID')['末次开票时间'].max()
last_bill_time_df = pd.DataFrame({'企业ID': last_bill_time.index, '末次开票时间': last_bill_time.values})

# 合并根据企业ID 合并 项目信息
result = pd.merge(dept_sale_class_df, open_bo_time_df, how='left', on='企业ID')
result = pd.merge(result, first_pay_time_df, how='left', on='企业ID')
result = pd.merge(result, last_pay_time_df, how='left', on='企业ID')
result = pd.merge(result, first_bill_time_df, how='left', on='企业ID')
result = pd.merge(result, last_bill_time_df, how='left', on='企业ID')
```

## KA区域项目-基于受票方-处理

```python
# 对KA区域下的项目 按 受票方 处理
data_ka_quyu.sort_values(by=['受票方名称','首次建单时间'],ascending=True,inplace=True)

# 获取每个受票方对应部门，销售，分级
spf_sale_class_df = data_ka_quyu.groupby('受票方名称')[['部门','销售','运营','联系人',
                                                  '联系电话','使用系统','服务费率','分级','业务类型']].last()
spf_sale_class_df.reset_index(inplace=True)
# 获取每个受票方最早开户时间最早建单时间最后建单时间
open_bo_time_df = data_ka_quyu.groupby('受票方名称').agg({'开户时间':'min','首次建单时间':'min','末次建单时间':'max'})
open_bo_time_df.reset_index(inplace=True)

# 剔除首次支付为空的项目
data_ka_quyu2 = data_ka_quyu.dropna(subset=['首次支付时间'])
# 获取每个项目最早支付时间
first_pay_time = data_ka_quyu2.groupby('受票方名称')['首次支付时间'].min()
first_pay_time_df = pd.DataFrame({'受票方名称':first_pay_time.index,'首次支付时间':first_pay_time.values})
# 获取每个项目最后支付时间
last_pay_time = data_ka_quyu2.groupby('受票方名称')['末次支付时间'].max()
last_pay_time_df = pd.DataFrame({'受票方名称':last_pay_time.index,'末次支付时间':last_pay_time.values})

# 剔除首次开票为空的项目
data_ka_quyu3 = data_ka_quyu2.dropna(subset=['首次开票时间'])
# 获取每个项目最早开票时间
first_bill_time = data_ka_quyu3.groupby('受票方名称')['首次开票时间'].min()
first_bill_time_df = pd.DataFrame({'受票方名称':first_bill_time.index,'首次开票时间':first_bill_time.values})
# 获取每个项目最后开票时间
last_bill_time = data_ka_quyu3.groupby('受票方名称')['末次开票时间'].max()
last_bill_time_df = pd.DataFrame({'受票方名称':last_bill_time.index,'末次开票时间':last_bill_time.values})

result2 = pd.merge(spf_sale_class_df,open_bo_time_df,how='left',on='受票方名称')
result2 = pd.merge(result2,first_pay_time_df,how='left',on='受票方名称')
result2 = pd.merge(result2,last_pay_time_df,how='left',on='受票方名称')
result2 = pd.merge(result2,first_bill_time_df,how='left',on='受票方名称')
result2 = pd.merge(result2,last_bill_time_df,how='left',on='受票方名称')
```

## 生成企业id的周建单数据 和 集团 企业ID 信息 合并

```python
# 获取当周周一日期
monday = datetime.date.today() + datetime.timedelta(days=-datetime.date.today().weekday())
```

```python
for i in range(10):
    monday_str = monday.strftime('%Y%m%d')
    last_monday = monday + datetime.timedelta(days=-7)
    last_monday_str = last_monday.strftime('%Y%m%d')
    week = day[(day['日期'] < monday_str) & (day['日期'] >= last_monday_str)].copy()
    week_qyid = week.groupby('企业ID')['建单金额'].sum()
    week_qyid_df = pd.DataFrame({'企业ID': week_qyid.index, 'week' + str(i + 1): week_qyid.values})
    monday = last_monday
    result = pd.merge(result, week_qyid_df, how='left', on='企业ID')
```

```python
result.info()
```

```python
a = result.columns.values.tolist()[:19]
b = result.columns.values.tolist()[19:]
c = a+list(reversed(b))
result = result[c]
```

```python
# 筛选前三周>0的项目
res = result.query('week2 > 0 and week3 > 0 and week4 > 0').copy()
res.fillna(0,inplace=True)
res.loc[res['week1'] / res[['week2', 'week3', 'week4']].mean(axis=1) > 0.2, '状态'] = '正常'
res.loc[res['week1'] / res[['week2', 'week3', 'week4']].mean(axis=1) <= 0.2, '状态'] = '预警'
```

## 导出集团-前三周有建单项目-周状态

```python
res.to_excel('集团-五月有建单项目中最后一周预警项目.xlsx',index=False)
```
